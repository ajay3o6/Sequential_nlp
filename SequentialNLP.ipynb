# Sentiment Classification


## Loading the dataset

```python
from keras.datasets import imdb

vocab_size = 10000 #vocab size

(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=vocab_size) # vocab_size is number of words to consider from the dataset, ordering based on frequency.
```

```python
(x_train.shape, y_train.shape), (x_test.shape, y_test.shape)
```

```python
import numpy as np

# concatenating the train and test data as both are of same length
data = np.concatenate((x_train, x_test), axis=0)
targets = np.concatenate((y_train, y_test), axis=0)

print("Categories:", np.unique(targets))
print("Number of unique words:", len(np.unique(data)))
```

```python
from keras.preprocessing.sequence import pad_sequences
vocab_size = 10000 #vocab size
maxlen = 300  #number of word used from each review
```

## Train test split

```python
# we split our data into training and testing set. The training set will contain 40,000 reviews and the testing set is 10,000.

x_train = data[10000:]
y_train = targets[10000:]

x_test = data[:10000]
y_test = targets[:10000]
```

```python
# #load dataset as a list of ints
# make all sequences of the same length
x_train = pad_sequences(x_train, maxlen=maxlen)
x_test =  pad_sequences(x_test, maxlen=maxlen)
```

```python
x_train.shape, y_train.shape,x_test.shape, y_test.shape
```

## Build Keras Embedding Layer Model
We can think of the Embedding layer as a dicionary that maps an index assigned to a word vector. The layer is flexible and is used in a few ways:

* The embedding layer can be used at the start of a large deep learning model. 
* We can load pre-train word embeddings into the embedding layer when the model is created.
* Using the embedding layer to train our own word2vec models.

The keras embedding layer doesn't require us to onehot encode our words, instead we have to give each word a unqiue intger number as an id. For the imdb dataset we've loaded this has already been done, but if this wasn't the case we could use sklearn [LabelEncoder](http://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.LabelEncoder.html).

```python
from keras.models import Sequential, Model
from tensorflow.keras.layers import Embedding, LSTM, Dense, Flatten, Dropout
```

```python
# Input - Layer
model = Sequential()

# Hidden - Layers (Embedding and LSTM layers)
model.add(Embedding(vocab_size, 100, input_length=maxlen))
model.add(LSTM(100))
model.add(Dropout(0.5))
# model.add(Flatten())
model.add(Dense(250, activation='relu'))


# Output- Layer
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

### Now we compile our model, which is nothing but configuring the model for training. We use the “adam” optimizer, an algorithm that changes the weights and biases during training. We also choose binary-crossentropy as loss (because we deal with binary classification) and accuracy as our evaluation metric.

```python
model.compile(
    optimizer="adam", 
    loss="binary_crossentropy", 
    metrics=["accuracy"])
```

### Now we're able to train our model. We'll do this with a batch_size of 64

```python
results = model.fit(x_train, y_train, validation_data=(x_test, y_test), batch_size=512, epochs=3) # Train
```

### It is time to evaluate our model:

```python
y_pred = model.predict_classes(x_test)
```

```python
import pandas as pd
# Comparision of sample predicted values with actuals
df = pd.DataFrame({'y_test': y_test[0:15], 'y_pred': y_pred[0:15].reshape(15,)})
df
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

### With this simple model we could able to achieve 88% of test accuracy which is Great! and further it can be improved by more epochs and tuning the hyperparameters along with the regularization techniques.

## Retrive the output of each layer in keras for a given single test sample from the trained model you built

```python
for i in range(0, len(model.layers)):
    tmp_model = Model(model.layers[0].input, model.layers[i].output)
    tmp_output = tmp_model.predict(x_test)[3]
    print(i+1, 'layer output\n')
    print(tmp_output,'\n','Size:',tmp_output.shape, '\n\n')
```
